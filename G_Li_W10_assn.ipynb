### TASK(1):  Create a .CSV file that includes all of the information. Load this .CSV file into GitHub repository. 

Try to read the CSV files and change it into a dataframe. 

```python
import pandas as pd
```

```python
#read the csv file
df = pd.read_csv("https://raw.githubusercontent.com/gegeli638/DAV-5400/master/week%2010%20data.csv")
df
```

```python
#put the csv file into a dataframe
flights = pd.DataFrame(df)
flights
```

### TASK(2):  Combining and reshaping data to tidy and transform the data.
Firstly, drop the NaN row in this dataframe. Use .drop function to achieve it.

```python
#drop the NaN value row
flights.drop(index=2, inplace=True)
flights
```

Secondly, fill the NaN into the airline name. Use .fillna to achieve it.

```python
#fill the NaN data
flights['Unnamed: 0'].fillna(method='ffill', inplace = True)
flights
```

Thirdly, rename the unnamed columns into 'airline' and 'type'. Use .rename function to achieve it.

```python
#rename the columns name
flights.rename(columns={"Unnamed: 0": "airline", "Unnamed: 1":"type"}, inplace = True)
flights
```

Finally, put this 'wide' format into 'long' format. Use the .melt function to achieve it.

```python
#put the 'wide' format into a 'long' format by using .melt function
flights_long = pd.melt(flights,['airline','type'])
flights_long.rename(columns={'variable':'city', 'value':'number'}, inplace=True)
flights_long
```

Now we can get the tidy data.
### TASK(3):  Answer the question: For each city, which airline had the best on time performance?
In order to answer this question, use a while loop to find all the raws in flights_long format. For the punctuality rate, use on time flights divided the sum of on time flights and delayed flights. In each city, which airline has the higher punctuality means it has the best on time performance.

```python
#use a while loop to find all the rows in the 'long' format
i = 0
while i <= len(flights_long)-1:
#calculate the punctuality rate in ALASKA airline
    ALASKA_Punctuality_rate = flights_long.iloc[i,3]/(flights_long.iloc[i,3] + flights_long.iloc[i+1,3])
#calculate the punctuality rate in AM WEST airline
    AMWEST_Punctuality_rate = flights_long.iloc[i+2,3]/(flights_long.iloc[i+2,3] + flights_long.iloc[i+3,3])
#print the result in direct words
    if ALASKA_Punctuality_rate>AMWEST_Punctuality_rate :
        print('In ' + flights_long.iloc[i,2] + ', ALASKA had the best on time performance')
    else:
        print('In ' + flights_long.iloc[i,2] + ', AM WEST had the best on time performance')
    i = i+4
```

### TASK(4): Transform the “long” format back into a “wide” format
I will transform the 'long' format into 'wide' format like the following example. The first index should be the airline, then type, then city. And then unstack the format, so we can see the on time flights and delayed flights in the same airline in each city. I think  this 'wide' presentation would be more easier to understand and work with.

```python
#use pivot_table function to change the index into 'airline', 'type' and 'city'
new = pd.pivot_table(flights_long, index=['airline','type','city'])
new
```

```python
#unstack the new format then show the 'wide' format
new.unstack().unstack()
```

In this 'wide' format, we can directly compare the on time and delayed flights in different airline for each city. It's more easier to show and answer the question: For each city, which airline had the best on time performance? This is an example of my recomendation.
