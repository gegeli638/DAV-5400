```python
import csv
 
with open('D:\programming/cars-sample35.txt') as csvfile:     
    readCSV = csv.reader(csvfile)
  
 #create new list
    price=[]
    cost=[]
    doors=[]
    passengers=[]
    luggage=[]
    safety=[]
    classification=[]
    
    list2=[]
    
    list3=[]
    
    list4=[]
    
    list5=[]
    
    list6=[]

    
 #first task 
 #use for loop to recreate seven new list
    for row in readCSV:
        price.append(row[0])
        cost.append(row[1])
        doors.append(row[2])
        passengers.append(row[3])
        luggage.append(row[4])
        safety.append(row[5])
        classification.append(row[6])
          
#second task
#use if to choose 'med' as a new list
        if row[0] == 'med':
            list2.append(row)
            
            
#third task
#after second tasl if, choose row[3](passengers) into a new list
            list3.append(row[3])
            
        
#fourth task
#use if to choose price 'high' and cost not 'low'
        if row[0]=='high' and row[1]!='low':
            list4.append(row)
           
        
#fifth task  
#use if to choose doors'2' and luggage 'big'
        if row[2]=='2' and row[4]=='big':
            list5.append(row)
   
 #final task
 #use for loop to change '5more' into '5', and change the type into int
    for a in doors:
        if a=='5more':
            a='5'
        a=int(a)
        list6.append(a)
 #def a new function average, rename it and use for loop to count the total number
    def average(total):
        s=0
        for b in total:
            s+=b
        return s/len(total)
   
 #print answers
    print(price, cost, doors, passengers, luggage, safety, classification)
    print(list2)
    print(list3)
    print(list4)
    print(list5)
    print(list6, average(list6))
    
    
    
 
```
